```python
import networkx as nx
```

```python
G = nx.Graph()
G.add_node(1)
G.add_node(2)
G.add_edge(1,2)
```

```python
G.add_nodes_from([3,4,5])
G.add_edges_from([(3,4),(3,5)])
```

```python
G.nodes()
G.edges()
G.neighbors(3)
nx.draw(G)
```

 - 노드와 링크 돌아보는 방법

```python
for n in G.nodes :
  print(n,G.neighbors(n))
for u, v in G.edges:
  print(u,v)
```

- 방향성 그래프 만들기

```python
D = nx.DiGraph()
D.add_edge(1,2)
D.add_edge(2,1)
D.add_edges_from([(2,3),(3,4)])
```

```python
D.number_of_nodes()
D.number_of_edges()
```

- 다양한 그래프 그리기

```python
D.neighbors(2)
D.predecessors(2)
D.successors(2)
nx.draw(D)
# 해당 노드를 향해 연결된 에지와 해당 노드로부터 나가는 에지만을 구분하는 함수
```

```python
B = nx.complete_bipartite_graph(4,5)
nx.draw(B)
```

```python
C = nx.cycle_graph(4)
nx.draw(C)
```

```python
S = nx.star_graph(6)
nx.draw(S)
```

```python
P = nx.path_graph(5)
nx.draw(P)
```
