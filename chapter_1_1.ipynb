```python
import networkx as nx
```

```python
G = nx.Graph()
G.add_node(1)
G.add_node(2)
G.add_edge(1,2)
```

```python
G.add_nodes_from([3,4,5])
G.add_edges_from([(3,4),(3,5)])
```

```python
G.nodes()
G.edges()
G.neighbors(3)
nx.draw(G)
```

 - 노드와 링크 돌아보는 방법

```python
for n in G.nodes :
  print(n,G.neighbors(n))
for u, v in G.edges:
  print(u,v)
```

- 방향성 그래프 만들기

```python
D = nx.DiGraph()
D.add_edge(1,2)
D.add_edge(2,1)
D.add_edges_from([(2,3),(3,4)])
```

```python
D.number_of_nodes()
D.number_of_edges()
```

### 가중치 그래프

```python
W = nx.Graph()
W.add_edge(1,2,weight=50)
W.add_weighted_edges_from([(2,3,3),(2,4,5)])
```

```python
nx.draw(W)
```

```python
for (i,j,w) in W.edges(data='weight') :
  if w > 3 :
    print('(%d, %d,%d' % (i,j,w))
```

```python
W.degree(2,weight='weight')
```

```python
print(nx.adjacency_matrix(G))
```

```python
print(nx.adjacency_matrix(D))
```

```python
print(nx.adjacency_matrix(W))
```

- 각 노드의 이웃을 출력하는 방법

```python
for n, neighbors in G.adjacency() :
  for number, link_attributes in neighbors.items():
    print('(%d, %d)'% (n,number))
```
